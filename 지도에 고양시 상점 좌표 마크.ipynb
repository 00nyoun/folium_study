```python
import pandas as pd
import numpy as np

df = pd.read_csv("./tobaccoRetail_goyang.csv",sep=",",encoding="euc-kr")
df
```

```python
df["WGS84위도"].isnull().sum()
```

```python
#좌표만 따로 df
df_coord = df[["WGS84위도","WGS84경도"]]
df_coord.head(3)
```

```python
df_coord.columns=["lat","lon"]
df_coord.head()
```

```python
df_coord
```

```python
#지도불러옴. 처음 위치
import folium
goyang_map=folium.Map(location=[37.6495,126.8089],zoom_start=12)
goyang_map
```

```python
#좌표를 지도에 마크 하기.
for i in range(len(df_coord)):
    shop = folium.Marker([df_coord.loc[i]["lat"],df_coord.loc[i]["lon"]],
                        icon=folium.Icon(color="blue"))
    shop.add_to(goyang_map)
```
